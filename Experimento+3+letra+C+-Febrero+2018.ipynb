PESO POSITIVO: AVGPOSVALENCE * AVGPOSAROUSAL

SUMA VALENCIA POSITIVA:  
SUMA ALERTA POSITIVA: 
AVGPOSITIVA: [SUMA VALENCIA POSITIVA/N(TOTAL DE PALAVRAS)* SUMA ALERTA POSITIVA/N (TOTAL DE PALAVRAS)]

    
PESO NEUTRAL: AVGNTRVALENCE * AVGNTRAROUSAL * REDUCING FACTOR

SUMA VALENCIA NEUTRA:
SUMA ALERTA NEUTRA:
AVGNEUTRA: [SUMA VALENCIA NEUTRA/N(TOTAL DE PALAVRAS)* SUMA ALERTA NEUTRA/N (TOTAL DE PALAVRAS) * 0.01]

    
PESO NEGATIVO: AVGNEGVALENCE * AVGNEGAROUSAL
    
SUMA VALENCIA NEGA:
SUMA ALERTA NEGA:
AVGNEGATIVA: [SUMA VALENCIA NEGATIVA/N(TOTAL DE PALAVRAS)* SUMA ALERTA NEGATIVA/N (TOTAL DE PALAVRAS)]

```python
#!/Users/imburguan/anaconda2/lib/python2.7/site-packages
# -*- coding: ISO-8859-1 -*-

import os, sys
import nltk
from Levenshtein import distance
from operator import itemgetter, attrgetter  


st1 = []
st2 = []
newbase =[]
palab =[]
conta = 0
vector = []
sum_valencia= 0 
#raices = 0
MenLD = 0
u = 1
vec = []
#b=[]
reload(sys)  
sys.setdefaultencoding('ISO-8859-1')

conta = 0
contpalpos = 0  
contpalntr = 0
contpalneg = 0

posval = 0
ntrval = 0
negval = 0

posarou = 0
ntrarou = 0
negarou = 0

avgposaro = 0
avgnegaro = 0
avgntraro = 0

#oracion = "Já Nebul desanimado conformado triste vida dia apareceu velhinha pedindo esmola"
oracion = "Mandou preparar nova esquadra quinhentos navios porque supunha filho morrera guerra travara reino Águas busca remédio cegueira"
#oracion = "Os dois irmãos aconselharam-lhe não continuasse viagem tempo perdido , pois país divertido , deixasse ficar ali"
#sno = nltk.stem.SnowballStemmer('portuguese')
pal = oracion.split()
print pal

for i in range(len(pal)):
	a = ""
	a= pal[i]
	print "ESTE ES EL VALOR DE i: ", i

	print len(vec)
	del vec[:]

	f=open("/Users/imburguan/Desktop/valencia4palabras.txt");
	for line in f:
		base=line.split()
		b = base[0]

		vct = 0
		print "pal[i].upper()  ", pal[i].upper()
		#print type(pal[i])
		print "b.upper()  ", b.upper()

		print "ESTE ES TAMANO VALOR DE VEC: ",  len(vec)
		

		if pal[i].upper() == b.upper():  
			vector.append(base)
			del vec[:]
			
			break
		else:
			MinLD = distance(pal[i].upper(), b.upper())
			lenpal = len (pal[i])

			letras1 = list(pal[i].upper())
			#print "letras1[0:3]   ", letras1[0:3] # 3 primeros caracteres de pal[i]

			letras2 = list(b)
			#print "letras2[0:3]   ", letras2[0:3] ## 3 primeros caracteres de pal[i]


			if ((MinLD <=2) and (lenpal >= 5) and (letras1[0:3] == letras2[0:3])):
				#print "La palavra de la frase con DL <= 2 : ", pal[i], b
				
				vec.append([base, MinLD]) 

				#print " Estos son los valores guardados que tienen distancia menor o igual a 2....  vec: ", vec

			else:

				stemmer = nltk.stem.RSLPStemmer()
				raiz1 = stemmer.stem(a)
				

				stemmer2 = nltk.stem.RSLPStemmer()
				raiz2= stemmer2.stem(b)
				
				print "MinLD:.......", MinLD

				if (raiz1.upper() == raiz2.upper() and (MinLD <= 2)):
					
					vector.append(base)

	print "len(vec) ################# : " , len(vec)
	if len(vec)	!= 0:

		for u in range(len(vec)):

			vec = sorted(vec, key=itemgetter(1)) #Ordeno ascendentemente 
		

		print "vec[0][-2]", vec[0][-2]
		sep = list(vec[0][-2])
		#print "sep: ", sep
		#print "sep[0]",sep[0]
		print type(sep[0])
		if sep[0] == pal[i].upper():
			vector.append(sep)
			
			break
		else:
	
			vector.append(sep)
			print "VALOR GUARDADO EN VECTOR ES: " , vector
	
		del vec[:]
	

	f.close()

for r in range(len(vector)):
	
	
	print "palavra do vetor: ", vector[r] ##palavra
	vector[r]= [float(h) for h in vector[r] if h != vector[r][0] ] #transformando a float excepto de la primera palabra

	print type (vector[r][0])

	if (float(vector[r][0]) >= 6):
		posval = posval +  float(vector[r][0])
		posarou = posarou + float(vector[r][1])
		print "Palavra positiva: " #, pal, vector[r][0]
		contpalpos = contpalpos + 1
	else:
		
		if (float(vector[r][0]) <= 4):
			negval = negval + float (vector[r][0])
			negarou = negarou + float (vector[r][1])
			print "Palavra negativa: " #, pal, vector[r][0]
			contpalneg = contpalneg + 1
		else:
			ntrval = ntrval + float (vector[r][0])
			ntrarou = ntrarou + float (vector[r][1]) 
			print "Palavra neutra: "#, pal, vector[r][0]
			contpalntr = contpalntr + 1
	conta = conta + 1

if conta ==0:
  promedio=0
  avgposval = 0
  avgntrval = 0
  avenegval = 0

else:
  
  avgposval = (posval/conta)
  avgposaro = posarou/conta

  if contpalneg ==0:
    avenegval=0
  else:

    avenegval = (negval/conta)
    avgnegaro = (negarou/conta)

  if contpalntr ==0:
    avgntrval=0
  else:

    avgntrval = (ntrval/conta)
    avgntraro = (ntrarou/conta)


print "###################"
print "avgposval: ", avgposval
print "avgposaro: ", avgposaro
peso_positivo= (avgposval) * (avgposaro)
print "PESO POSITIVO: avgposval * avgposaro: ", peso_positivo

print "###################"
print "avenegval: ", avenegval
print "avgnegaro: ", avgnegaro
peso_negativo = (avenegval) * (avgnegaro)
print "PESO NEGATIVO: avenegval * avgnegaro: ", peso_negativo

print "###################"
print "avgntrval: ", avgntrval
print "ntrarou: ", avgntraro
peso_neutro= (avgntrval) * (avgntraro)* 0.1 #Fator de reducção
print "PESO NEUTRO: avgntrval * avgntraro * 0.1: ", peso_neutro
print "###################"

print "O valor maximo é: ", max(peso_positivo, peso_negativo, peso_neutro)

```
